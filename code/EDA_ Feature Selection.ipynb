```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import Cleaned Data

```python
df = pd.read_csv('../datasets/clean_train.csv')
kaggle_test = pd.read_csv('../datasets/clean_test.csv')
```

# Look at correlations

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.corr()[['SalePrice']].sort_values('SalePrice', ascending=False), cmap='plasma', annot=True );
```

# Visualizations

```python
# making pairplots in subsets to choose variables
cols = df.columns
```

```python
start = 0
end = 5
for i in range(len(cols)+1):
    if end >= len(cols):
        end = len(cols)
    try:
        sns.pairplot(data=df,x_vars=df[cols[start:end]], y_vars='SalePrice')
    except: 
        pass
    start += 5
    end += 5
```

#### From the pairplots we can see that there are linear relationships. There are constant ones as well, since the columns are objects and not numerical
