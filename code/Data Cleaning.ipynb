```python
# Columns dropped for having too many null values - Alley, Fireplace Qu, Pool QC, 
# Fence, Misc Feature
df.drop(columns=['Alley', 'Fireplace Qu', 'Pool QC', 'Fence', 'Misc Feature'],
        axis=1, inplace=True)
kaggle_test.drop(columns=['Alley', 'Fireplace Qu', 'Pool QC', 'Fence', 
                          'Misc Feature'], axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
kaggle_test.isnull().sum()
```

```python
# dividing the variables into categorical and numerical and replacing null with 0, for 
# numerical columns and 'na' for categorical columns
def fill_missing_values(df, val_cat, val_num):
    for col in df.columns:
        if df[col].dtype in ['float64', 'int64']:
            df[col].fillna(val_num, inplace=True)
        else:
            df[col].fillna(val_cat, inplace=True)
```

```python
fill_missing_values(df, 0, 'na')
fill_missing_values(kaggle_test, 0, 'na')
```
